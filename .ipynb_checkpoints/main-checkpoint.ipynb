* EC Rating - Short for "early childhood," EC games are best fit for children under the age of 3.
* E Rating - Short for "everyone," E games are games that are good for people of all ages.
* T Rating - Short for "teen," T games are games that are best for anyone over the age of 13.
* M Rating - Short for "mature," M games are best for anyone over the age of 17.
* AO Rating - Short for "adults only," AO games are games that are best for anyone over the age of 18.
* RP Rating - Short for "rating pending," RP is games that have a rating that is still under investigation.

```python
fig, ax = plt.subplots(figsize=(8, 8))

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
    return my_autopct

genre_col = ['navy','crimson']
#genre_col = ['khaki','plum']

center_circle = plt.Circle((0, 0), 0.7, color='white')
plt.pie(x=console_count.values, labels=['PS4', 'PS4 & Xbox One'], autopct=make_autopct(console_count.values), 
          startangle=90, textprops={'size': 15}, pctdistance=0.5, colors=genre_col)
ax.add_artist(center_circle)

fig.suptitle('Number of Games for different Consoles', fontsize=20)
fig.show()
```

```python
f_req = df.columns[2:-1]
f_vector = np.array(df[f_req])

from sklearn.preprocessing import StandardScaler
stdSc = StandardScaler()
f_vector = stdSc.fit_transform(f_vector)

pca = PCA(n_components=2)
pca.fit(f_vector)
f_vector = pca.transform(f_vector)

dimReducedDataFrame = pd.DataFrame(f_vector)
dimReducedDataFrame['targets'] = df['esrb_rating']
dimReducedDataFrame = dimReducedDataFrame.rename(columns = {0:'V1', 1 : 'V2'})

plt.figure(figsize=(10, 10))
sns.scatterplot(data=dimReducedDataFrame, x='V1', y='V2', hue = 'targets')
plt.grid(True)
plt.show()
```

```python
X = df.drop(['title','esrb_rating'], axis=1)
y = df['esrb_rating']
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=40)
rf.fit(X_train, y_train)
```

```python
score = rf.score(X_train, y_train)
score2 = rf.score(X_test, y_test)
score, score2
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=200)

dt_clf.fit(X_train, y_train)

socre = accuracy_score(y_train, dt_clf.predict(X_train)) #training accuracy
score1 = accuracy_score(y_test, dt_clf.predict(X_test)) #testing accuracy

score, score1
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
model_params  = {
    "svm" : {
        "model":SVC(gamma="auto"),
        "params":{
            'C' : [1,10,20],
            'kernel':["rbf"]
        }
    },
    
    "decision_tree":{
        "model": DecisionTreeClassifier(),
        "params":{
            'criterion':["entropy","gini"],
            "max_depth":[5,8,9]
        }
    },
     "random_forest":{
        "model": RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10],
            "max_depth":[5,8,9]
        }
    },
    "naive_bayes":{
        "model": GaussianNB(),
        "params":{}
    },
    
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear',multi_class = 'auto'),
        'params': {
            "C" : [1,5,10]
        }
    }
    
}
```

```python
score=[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp["model"],mp["params"],cv=8,return_train_score=False)
    clf.fit(X,y)
    score.append({
        "Model" : model_name,
        "Best_Score": clf.best_score_,
        "Best_Params": clf.best_params_
    })
df_scores = pd.DataFrame(score,columns=["Model","Best_Score","Best_Params"])
```
